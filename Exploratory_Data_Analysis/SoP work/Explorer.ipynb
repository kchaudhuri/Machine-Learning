```python
random60 = pd.read_excel(file_name, sheet_name='60 random')
list_columns = list(random60.columns)
list_columns
```

```python
filtered_random60 = random60[['OBJECTID_1','LONGITUDE', 'LATITUDE', 'RAND()', 'ADDRESS found by KAB based on Long/Lat', 'WALKSCORE', 'TRANSSCORE', 'BIKESCORE']]
```

```python
filtered_random60 = filtered_random60.drop([60, 61])
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="kos_research")

def coordinatesToAddress(x):
    
    location = geolocator.reverse("{0}, {1}".format(x['LATITUDE'], x['LONGITUDE']))
    add = location.address
    print(add)
    return add
```

```python
filtered_random60["New Address"] = filtered_random60.apply(coordinatesToAddress, axis=1)
filtered_random60
```
