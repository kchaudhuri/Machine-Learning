```python
racedict = { 1 : 'black',
             2 : 'white',
             3 : 'latino',
             4 : 'asian',
             5 : 'native american',
             6 : 'other'}
```

```python
def raceindex(n):
    if n < 1 or n > 6:
        return racedict[0]
    return racedict[n]
```

```python
#applies function inside brace to the column
data['race'] = data['race'].apply(raceindex)

#applies and maps the dict values automatically
data['race_o'] = data['race_o'].map(racedict)
```
