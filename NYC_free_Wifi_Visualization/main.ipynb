```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('NYC_Free_Public_WiFi_03292017.csv')
```

```python
df.columns
```

```python
coords = df.as_matrix(columns = ['LAT', 'LON'])
```

```python
coords
```

```python
X=df.loc[:,['LAT','LON']]
```

```python
no_cl = 10
```

```python
kmeans = KMeans(n_clusters = no_cl, random_state = 1).fit(X)
klabel = kmeans.labels_
```

```python
ptsymb = np.array(['b.','r.','m.','g.','c.','k.','b*','r*','m*','r^']);
plt.figure(figsize=(12,12))
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
for i in range(no_cl):
    cluster=np.where(klabel==i)[0]
    plt.plot(X.LAT[cluster].values,X.LON[cluster].values,ptsymb[i])
plt.show()
```
